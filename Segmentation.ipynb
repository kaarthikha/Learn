# Scraping the data from the site and converting into dataframe

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from IPython.display import display_html


res = requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
soup = BeautifulSoup(res.content,'lxml')
table = soup.find_all('table')[0] 
df = pd.read_html(str(table))
dfs = df[0]
dfs.head()


```

# Filtering out Not assigned Borough

```python
df_borough=dfs[dfs.Borough != 'Not assigned']

df_borough.head()
```

```python
df_borough=df_borough.replace(' /', ',', regex=True)
df_borough.head()
```

```python
df_borough.shape
```

# Fetching the latitude and longitude and merging the dataframes


```python
df_cor=pd.read_csv('https://cocl.us/Geospatial_data')
df_cor.head()
```

```python
df_cor.columns=['0','Latitude','Longitude']
df_cor

df_borough.columns=['0','Borough','Neighborhood']
df_borough.head()
```

```python
df_inner = pd.merge(df_borough, df_cor, on='0', how='inner')
df_inner.columns=['Postal code','Borough','Neighborhood','Latitude','Longitude']
df_inner.head()
```

# Merged dataframe

```python
df_inner
```

# Fetching the Boroughs that contain Toronto

```python
toronto_data = df_inner[df_inner['Borough'].str.contains('Toronto')].reset_index(drop=True)
toronto_data.head()
```

```python
from geopy.geocoders import Nominatim # convert an address into latitude and longitude values
address = 'Downtown Toronto, Toronto'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

# Finding their neighborhood and plotting them on map

```python
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=11)

# add markers to map
for lat, lng, label in zip(toronto_data['Latitude'], toronto_data['Longitude'], toronto_data['Neighborhood']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
